```python
def find_correlated_columns(df, threshold=0.5):
    corr_matrix = df.corr().abs()  # Compute absolute correlation matrix

    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape,dtype = bool), k=1))

    # Find index of feature columns with correlation greater than threshold
    correlated_cols = [(col1, col2) for col1 in upper.columns for col2 in upper.columns
                       if upper[col1][col2] > threshold]

    return correlated_cols
```
